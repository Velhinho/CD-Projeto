```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np

HEIGHT: int = 4
NR_COL: int = 3

fp_ds1 = "./heart_failure_clinical_records_dataset.csv"
fp_ds2 = "./qsar_oral_toxicity.csv"
ds1 = pd.read_csv(fp_ds1, parse_dates=False, sep=',')
ds2 = pd.read_csv(fp_ds2, parse_dates=False, sep=';')
ds2.head()
ds2.dtypes
```

```python
columns = ds1.select_dtypes(include='number').columns
rows = len(columns)
bins = (10, 100)
cols = len(bins)
fig, axs = plt.subplots(rows, cols, figsize=(cols*HEIGHT + 4, rows*HEIGHT + 4))
for i in range(rows):
    for j in range(cols):
        axs[i, j].set_title('Histogram for %s %d bins'%(columns[i], bins[j]))
        axs[i, j].set_xlabel(columns[i])
        axs[i, j].set_ylabel('Nr records')
        axs[i, j].hist(ds1[columns[i]].values, bins=bins[j])
plt.show()
```
